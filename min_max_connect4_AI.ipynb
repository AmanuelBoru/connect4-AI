```python
################################
# this function returns a value of all the possible moves
################################
def possible_moves(board):
    all_list = []
    for i in range(len(board[0])):
        if board[-1][i] == EMPTY:
            all_list.append(i)
            
    return all_list
```

```python
################################
# this function allows the user or the bot to make a move
################################
def move(board, player, player_type):
    if player_type == 1:
        num = int(input("player " + player + " make a move: "))
        made = False
        index = 0


        while board[0-1][num - 1] != EMPTY:
            num = int(input("illegal move make, another move: "))
            
            

        while index < len(board) and made == False:
            if board[index][num - 1] == EMPTY:
                board[index][num - 1] = player
                made = True

            index += 1
            
    elif player_type == 2:
        possible = possible_moves(board)
        random.seed(datetime.now())
        value = randint(0, len(possible) - 1)
        add(board, player, possible[value])
    
    
    elif player_type == 3:
        next_move = minmax(board, player)
        add(board, player, next_move)
```

```python
################################
# this function adds a block to chosen space
################################
def add(board, player, column):
    made = False
    index = 0
    while index < len(board) and made == False:
        if board[index][column] == EMPTY:
            board[index][column] = player
            made = True
        index += 1
```

```python
################################
# this function makes removes a pice from the board
################################
def remove(board, column):
    index = len(board) - 1
    removed = False
    
    while index >= 0 and removed == False:
        if board[index][column] != EMPTY:
            board[index][column] = EMPTY
            removed = True
            
        index -= 1
```

```python
################################
# this function checks if a win can be found from thsi cell
################################
def win_cell(board, coordinate):
    winner = False
    
    y = coordinate[0]
    x = coordinate[1]
    y_length = len(board)
    x_length = len(board[0])
    
    value = board[y][x]
    
    # check upword
    if y <= y_length - 4:
        count = 0
        for i in range(4):
            if board[y + i][x] == value:
                count += 1
                
        if count == 4:
            winner = True
            

    #check downward
    if y >= 3:
        count = 0
        for i in range(4):
            if board[y - i][x] == value:
                count += 1
                
        if count == 4:
            winner = True
            
    # check forward
    if x <= x_length - 4:
        count = 0
        for i in range(4):
            if board[y][x + i] == value:
                count += 1
                
        if count == 4:
            winner = True
    
    # check forward and upowrd
    if x <= x_length - 4 and y <= y_length - 4:
        count = 0
        for i in range(4):
            if board[y + i][x + i] == value:
                count += 1

        if count == 4:
            winner = True

    # check forward and downward
    if x <= x_length - 4 and y >= 3:
        count = 0
        for i in range(4):
            if board[y - i][x + i] == value:
                count += 1

        if count == 4:
            winner = True
    
    
    # check backward
    if x >= 3:
        count = 0
        for i in range(4):
            if board[y][x - i] == value:
                count += 1
                
        if count == 4:
            winner = True
    
    # check downward backward
    if y >= 3 and x >= 3:
        count = 0
        for i in range(4):
            if board[y - i][x - i] == value:
                count += 1

        if count == 4:
            winner = True




    # check backward and upward
    if x >= 3 and y <= y_length - 4:
        count = 0
        for i in range(4):
            if board[y + i][x - i] == value:
                count += 1

        if count == 4:
            winner = True
                
    return winner
        
```

```python
################################
# this function checks if anyone has won the game
################################
def win_check(board):
    win = False
    
    i = 0
    while i < len(board) and win == False:
        j = 0
        while j < len(board[i]) and win == False:
            coordinate = [i, j]
            if board[i][j] != EMPTY:
                win = win_cell(board, coordinate)
                
            j += 1
        i += 1
                
    return win
        
```

```python
################################
# this function checks if a specific player has won the game
################################
def check_player_win(board, player):
    win = False
    
    i = 0
    while i < len(board) and win == False:
        j = 0
        while j < len(board[i]) and win == False:
            if board[i][j] == player:
                coordinate = [i,j]
                win = win_cell(board, coordinate)
                
            j += 1
        i += 1  
        
    return win
```

```python
################################
# this function checks if there are three in a row available
################################
def three_in_row(board, coordinate, player):
    row_total = 0
    
    y = coordinate[0]
    x = coordinate[1]
    y_length = len(board)
    x_length = len(board[0])
    
    value = player
    
    # check upword
    if y <= y_length - 4:
        count = 0
        for i in range(4):
            if board[y + i][x] == value:
                count += 1
                
        if count == 3:
            row_total +=1
            

    #check downward
    if y >= 3:
        count = 0
        for i in range(4):
            if board[y - i][x] == value:
                count += 1
                
        if count == 3:
            row_total +=1
            
    # check forward
    if x <= x_length - 4:
        count = 0
        for i in range(4):
            if board[y][x + i] == value:
                count += 1
                
        if count == 3:
            row_total +=1
    
    # check forward and upowrd
    if x <= x_length - 4 and y <= y_length - 4:
        count = 0
        for i in range(4):
            if board[y + i][x + i] == value:
                count += 1

        if count == 3:
            row_total +=1

    # check forward and downward
    if x <= x_length - 4 and y >= 3:
        count = 0
        for i in range(3):
            if board[y - i][x + i] == value:
                count += 1

        if count == 3:
            row_total +=1


    # check backward
    if x >= 3:
        count = 0
        for i in range(4):
            if board[y][x - i] == value:
                count += 1
                
        if count == 3:
            row_total +=1
    
    # check downward backward
    if y >= 3 and x >= 3:
        count = 0
        for i in range(4):
            if board[y - i][x - i] == value:
                count += 1

        if count == 3:
            row_total +=1




    # check backward and upward
    if x >= 3 and y <= y_length - 4:
        count = 0
        for i in range(4):
            if board[y + i][x - i] == value:
                count += 1

        if count == 3:
            row_total +=1
                
    return row_total
```

```python
################################
# this function checks if there are two in a row available
################################
def two_in_row(board, coordinate, player):
    row_total = 0
    
    y = coordinate[0]
    x = coordinate[1]
    y_length = len(board)
    x_length = len(board[0])
    
    value = player
    
    # check upword
    if y <= y_length - 3:
        count = 0
        for i in range(3):
            if board[y + i][x] == value:
                count += 1
                
        if count == 2:
            row_total +=1
            

    #check downward
    if y >= 2:
        count = 0
        for i in range(3):
            if board[y - i][x] == value:
                count += 1
                
        if count == 2:
            row_total +=1
            
            
    # check forward
    if x <= x_length - 3:
        count = 0
        for i in range(3):
            if board[y][x + i] == value:
                count += 1
                
        if count == 2:
            row_total +=1
    
    # check forward and upowrd
    if x <= x_length - 3 and y <= y_length - 3:
        count = 0
        for i in range(3):
            if board[y + i][x + i] == value:
                count += 1

        if count == 2:
            row_total +=1

    # check forward and downward
    if x <= x_length - 3 and y >= 2:
        count = 0
        for i in range(3):
            if board[y - i][x + i] == value:
                count += 1

        if count == 2:
            row_total +=1







    # check backward
    if x >= 2:
        count = 0
        for i in range(3):
            if board[y][x - i] == value:
                count += 1
                
        if count == 2:
            row_total +=1
        
    # check downward backward
    if y >= 2 and x >= 2:
        count = 0
        for i in range(3):
            if board[y - i][x - i] == value:
                count += 1

        if count == 2:
            row_total +=1


    # check backward and upward
    if x >= 2 and y <= y_length - 3:
        count = 0
        for i in range(3):
            if board[y + i][x - i] == value:
                count += 1

        if count == 2:
            row_total +=1
                
    return row_total
```

```python
################################
# this function calculates the players fitness
################################
def heuristic(board, player):
    oponent = PLAYER_ONE if player == PLAYER_TWO else PLAYER_TWO
    if check_player_win(board, player):
        return 500
    
    if check_player_win(board, oponent):
        return -500
    
    else:
        free_two = 0
        free_three = 0
        
        
        for i in range(len(board)):
            for j in range(len(board[i])):
                if board[i][j] == EMPTY:
                    coordinate = [i,j]
                    num = three_in_row(board, coordinate, player)
                    free_three += num
                    
                    num = two_in_row(board, coordinate, player)
                    free_two += num
        
        opponent_free_two = 0
        opponent_free_three = 0
        for i in range(len(board)):
            for j in range(len(board[i])):
                if board[i][j] == EMPTY:
                    coordinate = [i,j]
                    num = three_in_row(board, coordinate, oponent)
                    opponent_free_three -= num
                    
                    num = two_in_row(board, coordinate, oponent)
                    opponent_free_three -= num
        
        
        final_result = free_two + (free_three * 5) + opponent_free_two + (opponent_free_three * 5)
        
    return final_result
```

```python
################################
# this function recursively finds the fitness of all possible moves
################################
def rec_minmax(board, depth, player, maximizing):
    
    moves = possible_moves(board)
    oponent = PLAYER_ONE if player == PLAYER_TWO else PLAYER_TWO
    
    if maximizing:
        heuristic_value = heuristic(board, player)
    else:
        heuristic_value = heuristic(board, oponent)
        
        
    
    if heuristic_value == 500:
        
        return heuristic_value
    
    if heuristic_value == -500:
        
        return heuristic_value
    
    if len(moves) <= 0:
        
        return 0
    
    if depth <= 0:
        
        return heuristic_value
    
    if maximizing:
        best_value = -1000


        for i in range(len(board[0])):
            if i in moves:
                add(board, player, i)
                
                new_value = rec_minmax(board, depth - 1, player, not maximizing)
                remove(board, i)

                if new_value > best_value:
                    best_value = new_value
        
        return best_value
    else:
        best_value = 1000
        
        for i in range(len(board[0])):
            if i in moves:
                add(board, oponent, i)
                new_value = rec_minmax(board, depth - 1, player, not maximizing)
                remove(board, i)

                if new_value < best_value:
                    best_value = new_value
        
        return best_value
    
    ##################################
    
```
